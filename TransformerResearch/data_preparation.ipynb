```python
# Load the data from the file

import numpy as np
import pandas as pd

columns = ['par_id', 'art_id', 'keyword', 'country_code', 'text', 'label']
data = pd.read_csv('data/dontpatronizeme_pcl.tsv', sep='\t', encoding='utf-8', skiprows=4, names=columns, index_col=columns[0])
train_labels = pd.read_csv('data/train_labels.csv')
test_labels = pd.read_csv('data/test_labels.csv')
```

```python
train_labels
```

```python
# Merge the data with the labels just keeping the ones that are in the labels
train_data = data.merge(train_labels, how='right', on='par_id')
test_data = data.merge(test_labels, how='right', on='par_id')
train_data
```

```python
# if the label x is greater than 1 then it is a positive example
train_data['label_x'] = train_data['label_x'].apply(lambda x: 1 if x > 1 else 0)
test_data['label_x'] = test_data['label_x'].apply(lambda x: 1 if x > 1 else 0)
train_data
```

```python
# Save the data without the index
train_data.to_csv('data/train.csv')
test_data.to_csv('data/test.csv')
```
